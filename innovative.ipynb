```python
import numpy as np

f=open("data.txt","r")

f1=f.readline()
fit=f1.replace('\n',' ').split()

f2=f.readline()
st=f2.replace('\n',' ').split()

l1=len(fit)
l2=len(st)
print(fit)
print(st)

def firstfit():

  for i in range(l1):
      fit[i]=int(fit[i])

  for i in range(l2):
      st[i]=int(st[i])

  int_f=0
  ext_f=0
  fitb=[False]*l1
  stb=[False]*l2
  # print(fit[i])

  for i in range(l2):
      for j in range(l1):
          if fitb[j]==False and stb[i]==False:
              if fit[j]>=st[i]:
                  int_f+=fit[j]-st[i]
                  fitb[j]=True
                  stb[i]=True

  st1=[]
  fit1=[]
  print(int_f)
  for i in range(l2):
      if stb[i]==False:
          st1.append(st[i])
          
      
  for i in range(l1):
      if fitb[i]==False:
          fit1.append(fit[i])
    
  st1.sort()
  sum=0
  for i in range(len(fit1)):
    sum+=fit1[i]

  for i in range(len(st1)):
    if sum>=st1[i]:
      ext_f=st1[i]
      break

  print("Internal Fragmentation :",int_f)
  print("External Fragmentation :",ext_f)
  
def bestfit():
  for i in range(l1):
      fit[i]=int(fit[i])

  for i in range(l2):
      st[i]=int(st[i])

  int_f=0
  ext_f=0
  fitb=[False]*l1
  stb=[False]*l2
  # print(fit[i])

  fit.sort()
  for i in range(l2):
      for j in range(l1):
          if fitb[j]==False and stb[i]==False:
              if fit[j]>=st[i]:
                  int_f+=fit[j]-st[i]
                  fitb[j]=True
                  stb[i]=True

  st1=[]
  fit1=[]
  print(int_f)
  for i in range(l2):
      if stb[i]==False:
          st1.append(st[i])
          
      
  for i in range(l1):
      if fitb[i]==False:
          fit1.append(fit[i])
    
  st1.sort()
  sum=0
  for i in range(len(fit1)):
    sum+=fit1[i]

  for i in range(len(st1)):
    if sum>=st1[i]:
      ext_f=st1[i]
      break

  print("Internal Fragmentation :",int_f)
  print("External Fragmentation :",ext_f)


def worstfit():

  inf=0
  ext=0
  fitb=[False]*l1
  stb=[False]*l2

  fit.sort(reverse=True)
  for i in range(l2):
      for j in range(l1):
          if fitb[j]==False and stb[i]==False:
              if fit[j]>=st[i]:
                  inf+=fit[j]-st[i]
                  fitb[j]=True
                  stb[i]=True

  st1=[]
  fit1=[]
  print(inf)
  for i in range(l2):
      if stb[i]==False:
          st1.append(st[i])
          
      
  for i in range(l1):
      if fitb[i]==False:
          fit1.append(fit[i])
    
  st1.sort()
  sum=0
  for i in range(len(fit1)):
    sum+=fit1[i]

  for i in range(len(st1)):
    if sum>=st1[i]:
      ext=st1[i]
      break

  print("Internal Fragmentation :",inf)
  print("External Fragmentation :",ext)



while True:
  print("----------------------------------")
  print("|        1. First Fit            |")
  print("----------------------------------")
  print("|        2. Best Fit             |")
  print("----------------------------------")
  print("|        3. Worst Fit            |")
  print("----------------------------------")
  print("|        0. Exit                 |")
  print("----------------------------------")

  ch=int(input("Enter Choice : "))

  if ch==1:
    firstfit()
  elif ch==2:
    bestfit()
  elif ch==3:
    worstfit()
  else:
    break

    
```
